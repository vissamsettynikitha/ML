# Customer Segmentation with K-Means
This notebook segments customers using K-Means after preprocessing and visualization.
Steps: data loading, exploration, preprocessing, dimensionality reduction, clustering, and visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

Load the dataset and inspect the first few rows.

## Handle Missing Values
Remove rows with missing entries to simplify preprocessing.

```python
df = df.dropna()
print("Total values in the dataset after removing the null values:", len(df))
```

```python
df.nunique()
```

## Feature Engineering
Extract `day`, `month`, and `year` from `Dt_Customer`.

```python
parts = df["Dt_Customer"].str.split("-", n=3, expand=True)
df["day"] = parts[0].astype('int')
df["month"] = parts[1].astype('int')
df["year"] = parts[2].astype('int')
```

```python
df.drop(['Z_CostContact', 'Z_Revenue', 'Dt_Customer'],
        axis=1,
        inplace=True)
```

```python
floats, objects = [], []
for col in df.columns:
    if df[col].dtype == object:
        objects.append(col)
    elif df[col].dtype == float:
        floats.append(col)

print(objects)
print(floats)
```

## Encode Categorical Features
Label-encode categorical columns to make them suitable for clustering algorithms.

```python
for col in df.columns:
    if df[col].dtype == object:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
```

```python
plt.figure(figsize=(15, 15))
sb.heatmap(df.corr() > 0.8, annot=True, cbar=False)
plt.show()
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(df)
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2, random_state=0)
tsne_data = model.fit_transform(df)
plt.figure(figsize=(7, 7))
plt.scatter(tsne_data[:, 0], tsne_data[:, 1])
plt.show()
```

## Elbow Method to Select K
Run K-Means over a range of cluster counts and plot inertia to identify the elbow.

```python
error = []
for n_clusters in range(1, 21):
    model = KMeans(init='k-means++',
                   n_clusters=n_clusters,
                   max_iter=500,
                   random_state=22)
    model.fit(df)
    error.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 5))
sb.lineplot(x=range(1, 21), y=error)
sb.scatterplot(x=range(1, 21), y=error)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_scores = []
for k in range(2, 21):
    km = KMeans(init='k-means++', n_clusters=k, max_iter=500, random_state=22)
    labels = km.fit_predict(df)
    sil_scores.append(silhouette_score(df, labels, metric='euclidean'))
plt.figure(figsize=(10, 5))
sb.lineplot(x=range(2, 21), y=sil_scores)
sb.scatterplot(x=range(2, 21), y=sil_scores)
plt.ylabel('Silhouette score')
plt.xlabel('Number of clusters (k)')
plt.show()
```

## Fit K-Means
Train K-Means with the chosen number of clusters and obtain segment labels.

```python
model = KMeans(init='k-means++',
               n_clusters=6,
               max_iter=500,
               random_state=22)
segments = model.fit_predict(df)
```

```python
from sklearn.metrics import silhouette_score
final_sil = silhouette_score(df, segments, metric='euclidean')
print(f'Silhouette score (k=5): {final_sil:.4f}')
```

## Visualize Segments with t-SNE
Project data to 2D with t-SNE and color points by assigned segment to inspect separation.

```python
plt.figure(figsize=(7, 7))

df_tsne = pd.DataFrame({'x': tsne_data[:, 0], 'y': tsne_data[:, 1], 'segment': segments})

sb.scatterplot(x='x', y='y', hue='segment', data=df_tsne)
plt.show()
```
